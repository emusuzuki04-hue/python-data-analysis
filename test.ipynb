```python
import numpy as np
import matplotlib.pyplot as plt

# 1. データの準備（x軸）
# 0から2πまでの範囲を100分割したデータ（100個の点）を作成
x = np.linspace(0, 2 * np.pi, 100)

# 2. データの計算（y軸）
# xの値に対するsin関数（サインカーブ）の値を計算
y = np.sin(x)

# 3. グラフの描画
plt.figure(figsize=(8, 4))  # グラフのサイズを指定
plt.plot(x, y, label='sin(x)')  # x, y のデータを使ってグラフを描画し、凡例を設定

# 4. グラフの装飾
plt.title('Basic Sine Wave Graph')   # グラフのタイトル
plt.xlabel('X-axis (Radians)')      # X軸のラベル
plt.ylabel('Y-axis (sin(x))')       # Y軸のラベル
plt.grid(True)                      # グリッド（罫線）を表示
plt.legend()                        # 凡例を表示
plt.show()                          # グラフを表示
```
